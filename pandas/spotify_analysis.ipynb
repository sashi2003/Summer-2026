```python
import pandas as pd
import seaborn as sn
import numpy as np

df = pd.read_csv('spotify-tracks-dataset.csv')
print("Data loaded successfully...")
print("Shape: ",df.shape)
```

```python
df = df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
print("Cleaned. New shape: ",df.shape)
```

```python
df = df.dropna()
print("After dropping nulls", df.shape)
```

```python
df['duration_sec'] = df['duration_ms']/1000
print("Duration column added")
```

```python
print("Most Common Genres:")
print(df['track_genre'].value_counts().head(10))

print("\nAverage Populatiry",df['popularity'].mean().round(2))
print("Most Popular Track:")
print(df.loc[df['popularity'].idxmax(), ['track_name', 'artists', 'popularity']])
```

```python
genre_energy = df.groupby('track_genre')['energy'].mean()
genre_energy = genre_energy.sort_values(ascending = False)
print("Top 10 highest energy genres:", genre_energy.head(10))
print("Top 10 lowest energy genres:", genre_energy.tail(10))
```

```python
genre_dance = df.groupby('track_genre')['danceability'].mean()
genre_dance = genre_dance.sort_values(ascending = False)
print("Top 10 Danceable Tracks:",genre_dance.head(10))
print("\nBottom 10 Danceable Tracks:",genre_dance.tail(10))
```

```python
features = ['danceability', 'energy', 'loudness', 'speechiness', 
            'acousticness', 'instrumentalness', 'liveness', 
            'valence', 'tempo', 'duration_sec']
correlations = df[features].corrwith(df['popularity'])
correlations = correlations.sort_values(ascending = False)
print(correlations.round(3))
```

Surprisingly, no audio feature strongly predicts song popularity. The highest correlation is loudness at only 0.05, suggesting that popularity is driven by factors outside of audio characteristics such as artist fame, marketing, and cultural relevance. The strongest finding is that instrumental tracks tend to be less popular than vocal tracks.
